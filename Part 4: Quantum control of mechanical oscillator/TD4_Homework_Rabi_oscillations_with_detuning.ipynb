```python
# Rayan Trabelsi

import numpy as np
import matplotlib.pyplot as plt

V = 124e6  # coupling strength in Hz
tau_max = 40e-9  # maximum interaction time in seconds
delta_max = 150e6  # maximum detuning in Hz

tau = np.linspace(0, tau_max, 500)
delta = np.linspace(-delta_max, delta_max, 500)

T, D = np.meshgrid(tau, delta)

omega = np.sqrt(V**2 + D**2)

Pe = (V**2 / omega**2) * np.sin(omega * T / 2)**2

plt.figure(figsize=(10, 5))
plt.contourf(T*1e9, D*1e-6, Pe, levels=100, cmap='viridis')
plt.colorbar(label='Excited State Population $P_e$')
plt.xlabel('Interaction Time $\\tau$ (ns)')
plt.ylabel('Detuning $\\Delta$ (MHz)')
plt.title('Qubit Excited State Population vs Detuning and Interaction Time')
plt.axhline(y=0, color='red', linestyle='--', linewidth=1)
plt.show()
```
